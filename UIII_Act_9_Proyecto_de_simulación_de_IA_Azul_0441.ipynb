```python
# --- SISTEMA DE IA: MUSIC-MATCH ADVISOR ---

# Paso 1: Iniciamos con una lista maestra vacía para guardar los datos
base_datos_ia = []

# Variable para controlar la repetición del menú principal
ejecutando = "si"

# Iniciamos el bucle interactivo que simula el sistema
while ejecutando == "si":
    print("\n" + "="*40)
    print("      AI MUSIC SELECTOR 2026")
    print("="*40)
    print("1. Registrar cantante en la memoria IA")
    print("2. Ver biblioteca de artistas")
    print("3. Consultar recomendación de la IA")
    print("4. Salir")
    print("="*40)

    opcion = input("\nSelecciona una opción (1-4): ")

    # Opción 1: Agregar un nuevo cantante a la lista
    if opcion == "1":
        print("\n--- Registro de Artista ---")
        nombre = input("Nombre del cantante: ")
        genero = input("Género (Pop / Rock / Urbano): ").lower()
        ritmo = input("Energía (Lento / Enérgico): ").lower()

        # Guardamos los atributos en una sub-lista y la agregamos a la principal
        nuevo_perfil = [nombre, genero, ritmo]
        base_datos_ia.append(nuevo_perfil)
        print("\n¡Datos guardados con éxito en mi memoria!")

    # Opción 2: Mostrar los datos alineados como una tabla limpia
    elif opcion == "2":
        print("\n--- Biblioteca Musical de la IA ---")
        indice = 0
        total_registros = len(base_datos_ia)

        if total_registros == 0:
            print("La IA no tiene datos. Por favor, registra artistas.")
        else:
            # Imprimimos el encabezado de la tabla con alineación fija
            # (El número indica cuántos caracteres mide cada columna de ancho)
            print(f"{'CANTANTE':<20} | {'GÉNERO':<12} | {'RITMO':<12}")
            print("-" * 50) # Línea divisoria

            while indice < total_registros:
                perfil = base_datos_ia[indice]

                # Usamos f-strings para rellenar con espacios hacia la derecha (<)
                # Así, aunque el nombre sea largo o corto, las barras '|' se quedan quietas
                print(f"{perfil[0]:<20} | {perfil[1]:<12} | {perfil[2]:<12}")
                indice = indice + 1

    # Opción 3: Lógica de Inteligencia Artificial para recomendaciones
    elif opcion == "3":
        print("\nIA: ¡Hola! Analizaré tus gustos para darte el match ideal.")

        if len(base_datos_ia) == 0:
            print("IA: Error. Mi memoria está vacía. Necesito datos.")
        else:
            busqueda_gen = input("IA: ¿Qué género buscas hoy? ").lower()
            busqueda_rit = input("IA: ¿Buscas algo 'Lento' o 'Enérgico'? ").lower()

            # Recorremos la lista con un índice para buscar coincidencias
            i = 0
            encontrado = "no"

            print("\nIA: Procesando perfiles...")
            while i < len(base_datos_ia):
                perfil_actual = base_datos_ia[i]

                if perfil_actual[1] == busqueda_gen and perfil_actual[2] == busqueda_rit:
                    print("\n>>> ¡IA: RECOMENDACIÓN ENCONTRADA! <<<")
                    print("Te recomiendo escuchar a: " + perfil_actual[0])

                    if busqueda_gen == "rock":
                        print("Nota IA: ¡Este artista tiene guitarras potentes para ti!")
                    else:
                        print("Nota IA: Perfil musical verificado como óptimo.")

                    encontrado = "si"
                    i = len(base_datos_ia)
                else:
                    i = i + 1

            if encontrado == "no":
                print("\nIA: No encontré un perfil exacto con esos filtros.")

    # Opción 4: Finalizar el programa
    elif opcion == "4":
        print("\nIA: Apagando sistemas musicales. ¡Hasta luego!")
        ejecutando = "no"

    else:
        print("\nOpción no válida. Por favor, elige un número del 1 al 4.")

print("\n*** SESIÓN FINALIZADA ***")
```
